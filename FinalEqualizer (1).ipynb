```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import seaborn as sns
from scipy import stats
from scipy.optimize import curve_fit
```

```python
master_sheet = pd.read_csv("mastersheet.csv")
master_sheet
```

```python
plt.xlabel('Black')
plt.title('Teacher/Student Ratio')
plt.scatter(master_sheet["black%"], master_sheet['Student/Teacher'])
x = np.array(master_sheet["black%"])
y = np.array(master_sheet['Student/Teacher'])
res= stats.linregress(x,y)
plt.plot(x, res.intercept + res.slope*x,'r' )
plt.legend(["Linear Regression",'Actual Values'])
plt.show()
```
